# üå∏ Visualisasi Dataset Iris
Project ini berfokus pada eksplorasi visual terhadap dataset Iris.  
Tujuannya untuk memahami perbedaan fitur antar spesies dan melihat pola data menggunakan visualisasi dasar.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

df.head()
```

## üîç Pairplot
Visualisasi hubungan antar fitur berdasarkan masing-masing spesies bunga.

```python
sns.pairplot(df, hue='species', height=2)
plt.suptitle("Pairplot Iris Dataset", y=1.02)
plt.show()
```

## üìä Heatmap Korelasi
Menampilkan korelasi antar fitur numerik untuk melihat fitur mana yang paling saling berkaitan.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=['species']).corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap Korelasi Fitur Iris")
plt.show()
```

## üìà Distribusi Fitur
Visualisasi distribusi data per fitur berdasarkan jenis bunga.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='petal length (cm)')
plt.title("Distribusi Petal Length per Spesies")
plt.show()
```

## ‚úÖ Kesimpulan
- Petal length dan petal width sangat efektif untuk membedakan jenis bunga.
- Setosa memiliki ciri unik yang sangat berbeda dibanding dua lainnya.
- Visualisasi sangat membantu dalam memahami data sebelum membangun model ML.
